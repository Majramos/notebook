```python
localizacoes = []
for distrito in codigos:
    for concelho in codigos[distrito]:
        temp_morada = geocode(f'{concelho[0]}, {distrito}, portugal')
        
        localizacoes.append((temp_morada.latitude, temp_morada.longitude, concelho, distrito))
        
localizacoes[:5]
```

```python
map = folium.Map(location=(morada_galp.latitude, morada_galp.longitude),
                 zoom_start=8)
folium.Marker([morada_galp.latitude, morada_galp.longitude],
              popup='<i>Galp</i>',
              tooltip='Galp',
              icon=folium.Icon(color='orange')).add_to(map)

for loc in localizacoes:
    folium.Marker([loc[0], loc[1]], popup=f'<i>{loc[2]}, {loc[3]}, ({loc[0]}, {loc[1]})</i>').add_to(map)

map
```
